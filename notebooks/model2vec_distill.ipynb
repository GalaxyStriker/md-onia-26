```python
from model2vec.distill import distill

model_name = "Snowflake/snowflake-arctic-embed-l-v2.0"
pca_dims = 256
m2v_model = distill(
    model_name=model_name,
    pca_dims=pca_dims,
)
```

```python
model_local_path = f"../../models/snowflake_custom"
m2v_model.save_pretrained(model_local_path)
```

```python
from sentence_transformers import SentenceTransformer
from sentence_transformers.models import StaticEmbedding
from sentence_transformers.evaluation import NanoBEIREvaluator

static_embedding = StaticEmbedding.from_model2vec(model_local_path)
model = SentenceTransformer(modules=[static_embedding])

evaluator = NanoBEIREvaluator(
    show_progress_bar=True,
    batch_size=1,
    dataset_names=["hotpotqa"],
)

results = evaluator(model)
print(results)
print("Primary metric:", evaluator.primary_metric)
print("Primary score:", results[evaluator.primary_metric])
```
